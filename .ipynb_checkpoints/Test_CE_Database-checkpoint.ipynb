```python
import pandas as pd

df = pd.read_csv('Covid19_CasesTS_data.txt',sep='|')
new_columns = [column.replace(' ', '_').lower() for column in df]
df.columns = new_columns

!head Covid19_CasesTS_data.txt
!wc -l Covid19_CasesTS_data.txt
```

```python
from sqlalchemy.types import Integer, Text, String, DateTime, Numeric

df.to_sql("cases_ts",
               engine,
               if_exists='replace',
               schema='covid19',
               index=False,
               chunksize=50,
               dtype={"province": String(50),
                        "country": String(50),
                        "latitude": Numeric,
                        "longitude": Numeric,
                        "date_trunc": DateTime,
                        "confirmed": Integer,
                        "deaths": Integer,
                        "recovered": Integer,
                        "last_confirmed": Integer,
                        "new_cases": Integer,
                        "new_deaths": Integer
                });
df.head;
```
